```python
review_dates = reviews_dataframe['date'].tolist()
r_year = []
r_month = []
for i in reviews_dataframe['date'].index:
    y, m,d = reviews_dataframe['date'][i].split('-')
    r_year.append(y)
    r_month.append(m)
```

```python
reviews_dataframe['year'] = r_year
```

```python
reviews_dataframe = reviews_dataframe.rename(columns = {'user_id':'user_reviews'})
```

```python
reviews_dataframe.head()
```

```python
reviews_grouped = pd.DataFrame(reviews_dataframe.
                               groupby(['business_id','year','month'])
                               ['user_reviews'].apply(list))
reviews_grouped.reset_index()
```

```python
tips_grouped = pd.DataFrame(tips_dataframe.
                               groupby(['business_id','year','month'])
                               ['user_tips'].apply(list))
```

```python
tips_count = []
for i in tips_grouped.index:
    tips_count.append(len(tips_grouped['user_tips'][i]))
```

```python
tips_grouped['tips_count']=tips_count
```

```python
review_count = []
for i in reviews_grouped.index:
    review_count.append(len(tips_grouped['user_reviews'][i]))
reviews_grouped['review_count'] = review_count
```

```python
merged_tips_reviews = pd.merge(tips_grouped, reviews_grouped, how = 'outer',
                               on = ['business_id','year','month'],
                               right_index = True, left_index = False)
```

```python
merged_tips_reviews.head(n = 100)
```

```python
merged_tips_reviews['tip_count'] = ''
merged_tips_reviews['tip_user_count'] = ''
merged_tips_reviews['review_user_count'] = ''
merged_tips_reviews['overlap_user_count'] = ''

for i in merged_tips_reviews.index:
    
    merged_tips_reviews['tip_count'][i] = len(merged_tips_reviews['user_tips'][i])
    
    tip_set = set(merged_tips_reviews['user_tips'][i])
    review_set = set(merged_tips_reviews['user_reviews'][i])
    
    merged_tips_reviews['tip_user_count'][i] = len(tip_set)
    merged_tips_reviews['review_user_count'][i] = len(review_set)
    
    merged_tips_reviews['overlap_user_count'][i] = len(tip_set & review_set)

merged_tips_reviews.head()
```
