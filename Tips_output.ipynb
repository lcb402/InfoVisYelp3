```python
review_fields = ['business_id', 'stars','user_id']
reviews = pd.read_csv('yelp_academic_dataset_review.csv', skipinitialspace=True, usecols=review_fields)
reviews.head()
```

```python
ratings = reviews.groupby(['business_id'],as_index=False).mean()
ratings['rounded_stars'] = ratings['stars'].round()
ratings.head()
```

```python
review_users = pd.DataFrame(reviews.groupby('business_id')['user_id'].apply(list))
review_users = review_users.reset_index()
review_users = review_users.rename(columns = {'user_id':'review_user_ids'})
review_users.head()
```

```python
tip_users = pd.DataFrame(tips.groupby('business_id')['user_id'].apply(list))
tip_users = tip_users.reset_index()
tip_users = tip_users.rename(columns = {'user_id':'tip_user_ids'})
tip_users.head()
```

```python
business_data = pd.merge(tip_users, review_users, on='business_id')
business_data.head()
```

```python
business_data['tip_count'] = ''
business_data['tip_user_count'] = ''
business_data['review_user_count'] = ''
business_data['overlap_user_count'] = ''
for i in business_data.index:
    
    business_data['tip_count'][i] = len(business_data['tip_user_ids'][i])
    
    tip_set = set(business_data['tip_user_ids'][i])
    review_set = set(business_data['review_user_ids'][i])
    
    business_data['tip_user_count'][i] = len(tip_set)
    business_data['review_user_count'][i] = len(review_set)
    
    business_data['overlap_user_count'][i] = len(tip_set & review_set)

business_data.head()
```

```python
business_data = pd.merge(business_data, ratings, on='business_id')
business_data.head()
```
